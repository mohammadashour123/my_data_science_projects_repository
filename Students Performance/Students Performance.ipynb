## Data Exploration 

```python
plt.figure(figsize = (15, 8 ))
sns.histplot(data = data , x = 'reading score')
plt.title(' reding score plot \n')
```

```python
plt.figure(figsize = (15, 8 ))
sns.boxplot(data=data,x='math score')
```

### Function to display the distribution of data using  histplot and boxplot

```python
def plot_data(col):
    #plt.figure(figsize = (15, 8 ))
    fig , ax= plt.subplots(1 ,2  , figsize = (15, 8 ))
    sns.histplot(data = data , x = col , ax=ax[0])
    ax[0].set_title(f'{col}  score plot \n')
    
    sns.boxplot(data = data , x = col , ax=ax[1])
    ax[1].set_title(f'{col}  score plot \n')
```

```python
plot_data('math score')
```

```python
plot_data('reading score')
```
