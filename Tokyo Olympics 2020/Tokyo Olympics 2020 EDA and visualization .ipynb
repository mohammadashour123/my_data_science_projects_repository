## Details
This contains the details of over 11,000 athletes, with 47 disciplines, along with 743 Teams taking part in the 2021(2020) Tokyo Olympics.
This dataset contains the details of the Athletes, Coaches, Teams participating as well as the Entries by gender. It contains their names, countries represented, discipline, gender of competitors, name of the coaches.

Will update the dataset with medals(gold, silver, bronze), more details about the athletes after few weeks.

## let's descover top 10 Top Countries with most number of medals

```python
Top_Countries_with_most_number_of_medals = medals_data['Total'].head(10)
```

```python
plt.figure(figsize=(10,12))
my_labels = medals_data['Team/NOC'].head(10)
plt.pie(data  = medals_data , x = Top_Countries_with_most_number_of_medals , labels=my_labels , autopct='%1.1f%%') 
plt.title('Top 10 Countries with most number of medals', fontsize = 25)
plt.axis('equal');
```

### let's descover top 10 Top Countries with most number of gold medals

```python
# Create a datafeame included team and how much metal gold have 
gold_medals = medals_data[['Team/NOC' , 'Gold']]

# sort values by gold column ascending 
gold_medals = gold_medals.sort_values(['Gold'] , ascending=False)

# save the top results in variable 
gold_medals = gold_medals.head(10)

# display it 
gold_medals
```

```python
plt.figure(figsize=(10,12))
my_labels_of_gold = gold_medals['Team/NOC']
plt.pie(data  = gold_medals  , x= gold_medals['Gold'] , labels=my_labels_of_gold , autopct='%1.1f%%') 
plt.title('Top 10 Countries with most number of gold medals', fontsize = 25)
plt.axis('equal');
```

### let's descover top 10 Top Countries with most number of Silver medals

```python
# Create a datafeame included team and how much team have Silver medals 
silver_medals = medals_data[['Team/NOC' , 'Silver']]

# sort values by silver column ascending 
silver_medals = silver_medals.sort_values(['Silver'] , ascending=False)

# save the top results in variable 
silver_medals = silver_medals.head(10)

# display it 
silver_medals
```

#### let's make it more visabl ! 

```python
plt.figure(figsize=(12,15))
my_labels_of_silver = silver_medals['Team/NOC']
plt.pie( data = silver_medals , x = silver_medals['Silver'] , labels=my_labels_of_silver  ,autopct='%1.1f%%'  );
```

### let's descover top 10 Top Countries with most number of bronz medals

```python
# Create a datafeame included team and how much team have Bronze medals 
bronze_teams = medals_data[['Team/NOC' , 'Bronze']]

# sort values by Bronze column ascending 
bronze_teams = bronze_teams.sort_values(['Bronze'] , ascending=False)  

# save the top results in variable 
bronze_teams = bronze_teams.head(10)
```

```python
plt.figure(figsize=(12,15))
my_labels_of_bronze = bronze_teams['Team/NOC']
plt.pie(data = bronze_teams , x = bronze_teams['Bronze'] ,labels=my_labels_of_bronze ,  autopct='%1.1f%%' ) 
plt.title('Top 10 Countries with most number of Bronze medals ' , fontsize = 25 ) ; 
```

## HIGHEST NUMBER OF COACHES

```python
coaches_data.head()
```

```python
num_of_coaches = coaches_data.groupby('NOC').size()
num_of_coaches = num_of_coaches.to_frame('count')
num_of_coaches = num_of_coaches.sort_values('count' , ascending=False).head(10)
plt.figure(figsize=(12 , 10 ))
plt.bar(x = num_of_coaches.index.values , height=num_of_coaches['count'] , width=0.6)
plt.xlabel('Countries' , fontsize = 20)
plt.ylabel('Coach Count', fontsize = 20)
plt.xticks(rotation=45) ;
```
